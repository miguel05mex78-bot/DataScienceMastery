### Random Forest Regressor

#### Diabetes Dataset
- https://www.kaggle.com/uciml/pima-indians-diabetes-database
- The Pima Indians Diabetes Database is a collection of medical records of Pima Indians living in Arizona, USA.
- The dataset contains 442 observations and 10 variables.
- The variables are:
  - age     age in years
  - sex
  - bmi     body mass index
  - bp      average blood pressure
  - s1      tc, T-Cells (a type of white blood cells)
  - s2      ldl, low-density lipoproteins
  - s3      hdl, high-density lipoproteins
  - s4      tch, thyroid stimulating hormone
  - s5      ltg, lamotrigine
  - s6      glu, blood sugar level

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
diabetes = datasets.load_diabetes()
# it whows the clues (name of the elements) in the diabetes object
diabetes.keys()

```

```python
# prints the description of the dataset stored in diabetes.
print(diabetes['DESCR'])
```

```python
# separates the dataset into input features and target values.
X = diabetes.data
y = diabetes.target
```

```python
# Splitting the dataset
# X → input features
# y → target values
# 📊 Outputs:
# X_train → training features (70% of data)
# X_test → testing features (30% of data)
# y_train → training labels
# y_test → testing labels
# ⚙️ Parameters:
# test_size=0.3 → 30% of data is used for testing
# random_state=42 → ensures the split is reproducible (same result every time)

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=42)
# This creates a scaling object that will standardize your data.
scaler = StandardScaler()
# Two things happen here:
# fit → learns the mean and standard deviation from X_train
# transform → scales the data using that information
# 📌 After this, your features have:
# mean ≈ 0
# standard deviation ≈ 1
X_train = scaler.fit_transform(X_train)
# 👉 This scales the test data using the same values learned from training data.
# ⚠️ Important:
# You do NOT use fit_transform here
# This avoids data leakage

X_test = scaler.transform(X_test)
```

```python
pd.DataFrame(X_train).describe()
```

### Regression Model Building

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
```

```python
rfr = RandomForestRegressor(random_state=42)
rfr.fit(X_train, y_train)
y_predict = rfr.predict(X_test)
```

```python
# RMSE
np.sqrt(mean_squared_error(y_test, y_predict))
```

```python
# r2_score
r2_score(y_test, y_predict)
```

```python
# Hyperparameter Tuning
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'n_estimators': [100, 200, 400],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 10, 100],
    'min_samples_leaf': [1, 5, 10]}

rfr = RandomForestRegressor()
grid = GridSearchCV(estimator=rfr, param_grid=params, cv=5, n_jobs=-1)
grid.fit(X_train, y_train)
```

```python
grid.best_estimator_
```

```python
y_predict = grid.predict(X_test)
```

```python
# RMSE
np.sqrt(mean_squared_error(y_test, y_predict))
```

```python
# r2_score
r2_score(y_test, y_predict)
```
